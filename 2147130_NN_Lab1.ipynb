#2. UNIT Step

```python
x = np.arange(-20,20,1)
z= []
for i in x:
  if i < 0:
    z.append(0)
  elif i == 0:
    z.append(0.5)
  else:
    z.append(1)

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("unit step(x)")
plt.show()
```

#3. Linear

```python
x = np.arange(-100,100,2)
z= x

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Linear(x)")
plt.show()
```

*** 4. SIGN (SIGNUM)***

```python
x = np.arange(-20,20,1)
z= []
for i in x:
  if i < 0:
    z.append(-1)
  elif i == 0:
    z.append(0)
  else:
    z.append(1)

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Sign(x)")
plt.show()

```

# **5. Piece wise linear **

```python
x = np.arange(-30,30,1)
z= []
for i in x:
  if i >= 1/2:
    z.append(1)
  elif i > -1/2 and i < 1/2:
    z.append(i +(1/2))
  else:
    z.append(0)

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Piece Wise Linear(x)")
plt.show()

```

# **6. Hyperbolic Tangent**

```python
x = np.arange(-20,20,1)
z = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Hyperbolic(x)")
plt.show()
```
