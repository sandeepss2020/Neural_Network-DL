```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Iris_cat.csv')
dataset.isnull()
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
Le = LabelEncoder()
```

```python
dataset['Species'] = Le.fit_transform(dataset['Species'])
```

```python
dataset.info()
```

```python
features = dataset[['SepalLengthCm', 'PetalLengthCm']]
label = dataset[['Species']]
# print(features, label)
X = dataset[['SepalLengthCm', 'PetalLengthCm']]
Y = dataset[['Species']]
```

```python
x_train, x_test ,y_train, y_test = train_test_split(X, Y, test_size=0.6, shuffle=True, stratify=Y)
```

```python
classifier = MLPClassifier(hidden_layer_sizes=(4,2), max_iter=300, activation ='relu', solver='adam', random_state=1)
classifier_log = MLPClassifier(hidden_layer_sizes=(4, 2), max_iter=300, activation ='logistic', solver='adam', random_state=1)
```

```python
print(x_train, "\n", y_train)
```

```python
classifier.fit(x_train, y_train)
classifier_log.fit(x_train, y_train)
```

```python
print(f"for relu : {classifier.coefs_}")
print(f"for Logistic : {classifier_log.coefs_}")
```

```python
y_pred = classifier.predict(x_test)
y_pred_log = classifier_log.predict(x_test)
```

```python
print(confusion_matrix(y_pred, y_test))

print(confusion_matrix(y_pred_log, y_test))
```

```python
print("for relu : ",accuracy_score(y_pred, y_test))
print("for logistic : " ,accuracy_score(y_pred_log, y_test))
```
