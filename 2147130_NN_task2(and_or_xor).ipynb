```python
# importing Python library
import numpy as np
```

```python
# define Unit Step Function
def actual_result(v):
    if v > 1:
        return 1
    else:
        return 0
  
# design Perceptron Model
actual=[]
def perceptronModel(x, w, b):
    v = np.dot(w, x) + b
    y = actual_result(v)
    print(v)
    actual.append(v)
    return y
  
# AND Logic Function
# w1 = 1.5, w2 = 0.5, b = -0.5
def AND_logicFunction(x):
    w = np.array([1.5, 0.5])
    b = -0.5
    return perceptronModel(x, w, b)



```

```python
# OR Logic Function
# w1 = 1.5, w2 = 0.5, b = -0.5
def OR_logicFunction(x):
    w2 = np.array([1.5, 0.5])
    b2 = -0.5
    return perceptronModel2(x, w2, b2)


# design Perceptron Model
actual2=[]
def perceptronModel2(x, w2, b2):
    v2 = np.dot(w2, x) + b2
    y2 = actual_result2(v2)
    print(v2)
    actual2.append(v2)
    return y2

# define Unit Step Function
def actual_result2(v2):
    if v2 >= 0:
        return 1
    else:
        return 0
```

```python
def XOR_logicFunction(x):
    w3 = np.array([1.5, 0.5])
    b3 = -0.5
    return perceptronModel3(x, w3, b3)

actual3=[]
def perceptronModel3(x, w3, b3):
    v3 = np.dot(w3, x) + b3
    y3 = actual_result3(v3)
    print(v3)
    actual2.append(v3)
    return y3

# define Unit Step Function
def actual_result3(v3):
    if v3 >= 0:
        return 1
    else:
        return 0
```

```python
# testing the Perceptron Model
test1 = np.array([0, 0])
test2 = np.array([0, 1])
test3 = np.array([1, 0])
test4 = np.array([1, 1])
```

```python
def AND_operation():
  print("Actual output for perceptron\n")
  print("AND({}, {}) = {}".format(0, 0, AND_logicFunction(test1)))
  print("AND({}, {}) = {}".format(0, 1, AND_logicFunction(test2)))
  print("AND({}, {}) = {}".format(1, 0, AND_logicFunction(test3)))
  print("AND({}, {}) = {}".format(1, 1, AND_logicFunction(test4)))

  actual_out = [0,0,0,1]
  print("\Target Value\n")
  print("AND({}, {}) = {}".format(0, 0, actual_out[0]))
  print("AND({}, {}) = {}".format(0, 1, actual_out[1]))
  print("AND({}, {}) = {}".format(1, 0, actual_out[2]))
  print("AND({}, {}) = {}".format(1, 1, actual_out[3]))

  print("\nError Values : \n")
  total_error = 0
  for i in range(len(actual)):
    error = actual_out[i] - actual[i]
    print(error)
    total_error += error
  print("Total Error :- ",total_error)
  
```
