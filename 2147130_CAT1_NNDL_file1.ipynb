```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
features = np.array(
	[
		[1, 1],
		[1, -1],
		[-1, 1],
		[-1, -1]
	])

labels = np.array([1, 1, 1, -1])
print(f"Orginial Matrix:\n {features} \n Target values : \n{labels}")

# initialising the weights, bias , learning rate, epoch values
weight = [0.1, 0.1]
bias = 0.1
learning_rate = 0.1
epoch = 5

for i in range(epoch):
    print("epoch :", i+1)
    sum_squared_error = 0.0       #sum of squared error
    

    # for each of the possible input given in the features
    for j in range(features.shape[0]):
      actual = labels[j]
      x1 = features[j][0]
      x2 = features[j][1]
      unit = (x1 * weight[0]) + (x2 * weight[1]) + bias
      # summation of squared error is calculated
      error = actual - unit
      sum_squared_error += error * error
      print("error =", error)
      if error != 0:
        weight[0] += learning_rate * error * x1
        weight[1] += learning_rate * error * x2
        bias += learning_rate * error
    print("sum of squared error = ", sum_squared_error/4, "\n\n")
   

      
```
